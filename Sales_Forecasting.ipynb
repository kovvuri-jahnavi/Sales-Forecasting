```python
# Import Libraries

import pandas as pd
import numpy as np

print("Libraries Imported Successfully")
```

```python
# Load Dataset

df = pd.read_csv("sales_data.csv")

df
```

```python
# Create Numerical Month Values

df["Month_Number"] = range(1, len(df)+1)

df
```

```python
# Features and Target

X = df[["Month_Number"]]

y = df["Sales"]

print(X.head())
print(y.head())
```

```python
# Train Model

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, y)

print("Model Trained Successfully")
```

```python
# Predict Future Sales

future_months = pd.DataFrame({
    "Month_Number":[13,14,15]
})

predictions = model.predict(future_months)

print("Future Sales Predictions")
print(predictions)
```

```python
# Create Forecast Report

forecast_df = pd.DataFrame({
    "Month_Number":[13,14,15],
    "Predicted_Sales":predictions
})

forecast_df
```

```python
# Save Forecast Report

forecast_df.to_csv(
    "sales_forecast_report.csv",
    index=False
)

print("Forecast Report Saved Successfully")
```

```python
# Visualization

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(
    df["Month_Number"],
    df["Sales"],
    marker="o"
)

plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```
